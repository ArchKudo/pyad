## Algorithms

### What is an algorithm?
*From Wikipedia,* an algorithm is an unambigous specification of how to solve a class of problems.


### Properties of algorithms
1. It is an __effective method__ expressed in finite amount of __space__ and __time__.
2. Can be expressed in a formal language to calculate a mathematical function.
3. Has an initial [state](https://en.wikipedia.org/wiki/State_(computer_science))(event or user interaction) and input.
4. Transition from state(remberance of preceding events) may be either
    - [deterministic](https://en.wikipedia.org/wiki/Deterministic_system) i.e produce same output for given initial state, 
    - or can be random.

### Effective Methods
[Effective Methods](https://en.wikipedia.org/wiki/Effective_method) work on specific class of problems and have the following properties:
* Consists of __finite__ number of __instructions__.  
* When applied to a class of problems,
    - __Finishes__ after a finite number of steps.
    - Always produces a __right__ answer.
* Can be written down without any aids.
* Requires no ingenuity to succeed and  can be carried out __rigorous__ly.  

## Analyzing Algorithms

### An example of algorithm to calculate the sum of first `n` natural number

```python
def nthsum(n):
    """Calculate sum of n natural numbers using for-loop"""
    return sum([x for x in range(1, n+1)])     
```

```python
nthsum(10)
```

```python
def gausssum(n):
    """Calculate sum using gauss formula"""
    return n * (n+1) // 2 # Floor division, returns int
```

```python
gausssum(10)
```

### Timing both the algorithms

```python
%timeit nthsum(1000)
```

```python
%timeit gausssum(1000)
```

```python
%timeit nthsum(10000)
```

```python
%timeit gausssum(10000)
```

__NOTE__: 
* The `gausssum` function takes approximately __constant__ time for calculating the sum of 1000 or 10000 numbers.  
* However, the time required by `nthsum` increases for calculating 10000 as compared to 1000 numbers i.e the function __grows with input__.

### Graphically

```python
from numpy import mean
import timeit
import matplotlib.pyplot
```

```python
def nthsum(n):
    """Calculate sum of n natural numbers using for-loop"""
    return sum([x for x in range(1, n + 1)])


def gausssum(n):
    """Calculate sum using gauss formula"""
    return n * (n + 1) // 2  # Floor division, returns int
```

```python
def itertime(func, module='__main__', start=100, end=10000, step=100, repeat=10):
    """Return a list of mean time required for iteration of a function once"""
    meantimes = list()
    for i in range(start, end, step):
        # String of function call to be used in timeit.repeat
        fnstr = f'{func.__name__}({i})'
        
        # Import function for timeit.repeat
        mainstr = f'from {module} import {func.__name__}'
        
        # A lst of time required to call the function once repeated `repeat` times
        timelst = list(timeit.repeat(fnstr, mainstr, repeat=repeat, number=1))
        
        # Append mean of timelst for given `i`
        meantimes.append(mean(timelst))
    return meantimes
```

```python
itertime(nthsum)
```

```python
itertime(gausssum)
```

```python
def comp_graph():
    """Plot nthsum and gaussum"""
    nthsum_lst_in_us = [x * 10**6 for x in itertime(nthsum)]
    gaussum_lst_in_us = [x * 10**6 for x in itertime(gausssum)]
    inputsize = [x for x in range(len(nthsum_lst_in_us))] # Since nthsum and gausssum lists are of equal length
    
    matplotlib.pyplot.xlabel('Input Size(10**x)')
    matplotlib.pyplot.ylabel('Time required')
    
    matplotlib.pyplot.plot(inputsize, nthsum_lst_in_us) # Plot nthsum
    matplotlib.pyplot.plot(inputsize, gaussum_lst_in_us) # Plot gausssum
    
    matplotlib.pyplot.legend(['nthsum', 'gausssum'])
    
    matplotlib.pyplot.show()
```

```python
comp_graph()
```

Here we can see the `nthsum` (approximately) grows __linearly__ in time required for calculation of sum with increase input size,  
whereas the `gaussum` function is __constant__ with increase in input size.

## Time v/s Space Complexity

### Time Complexity
The time complexity of an algorithm quantifies the amount of time taken by an algorithm as a function of size of its input.

### Space Complexity
It represents the total amount of memory space that a normal(physical) computer would require to solve a problem using the algorithm.

## Asymptotic Analysis
Asymptotic Analysis analyzes algorithm's running time by identifying its behvaiour as the input size of the algorithm increases(i.e the algorithmic growth rate).  

### Alternatives
1. Counting number of primitive operation in an algorithm at different imput sizes.
    - Valid solution but can be tedious
2. Measure the amount of time required by an algorithm for differnt sizes. (As seen above)
    - Can be inaccurate and Machine-dependent(processing power, environment variables, architecture) etc.

```python
# For alternative #2
comp_graph() # Under light load
```

```python
comp_graph() # Under medium load
```

```python
comp_graph() # Under heavy load
```

It can be seen that time required isn't the same in all the graphs even for the same function,  
Thus one cannot __accurately__ measuring __performance__ of an algorithm by calculating the time required.

### Growth rate of mathematical functions

Some important functions:
0. Constant
1. Logarithmic (base 2, e, 10)
2. Linear
3. Quadratic
4. Polynomial
5. Exponential
6. Factorial

### Visualizing functions

```python
from math import log, log2, log10, factorial
```

```python
def vis_fngrowth():
    fnl = [lambda x: 1, lambda x: log2(x), lambda x: log(x), lambda x: log10(x), lambda x: x,
            lambda x: x**2, lambda x: x**3,
           lambda x: 3**x, lambda x: factorial(x)]
    
    fnliter = [[fn(x) for x in range(1,100) if fn(x) < 1000] for fn in fnl]
    matplotlib.pyplot.subplot()
    for lst in fnliter:
        matplotlib.pyplot.plot([x for x, _ in enumerate(lst)] , lst)
        matplotlib.pyplot.xticks(range(0, 100, 5))
        matplotlib.pyplot.yticks(range(1, 1000, 50))
        matplotlib.pyplot.legend(['Constant', 'log2', 'loge', 'log10', 'Linear', 'Quadratic', 'Polynomial',
                                 'Exponential', 'Factorial'])
    matplotlib.pyplot.show()
```

## Big-O Notation
Big-O or __O__ provides asymptotic __upper bound__ for the growth rate of runtime of a function.  
It is an asymptotic notation for __worst case__ or __ceiling__ of growth for a given function.
Say we have an algorithmic runtime of f(n), and g(n) is any arbitrary time complexity.  
We have $f(n) = O(g(n))$, if  $f(n) <= c g(n)$, for some real constant $c (c > 0)$, and for every input size $n (n > n_0)$.
