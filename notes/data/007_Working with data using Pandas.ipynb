## Reading HTML

```python
css_selectors = pd.read_html('https://www.w3schools.com/cssref/css_selectors.asp')
```

```python
type(css_selectors)
```

```python
css_df = pd.DataFrame({'Selectors': css_selectors[0][0][1:],
                      'Example': css_selectors[0][1][1:],
                      'Description': css_selectors[0][2][1:]})
```
